```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup  # This will parse out the HTML data 
import re
from datetime import datetime
import pandas as pd
```

```python

# Define NYC MTA turnstile data URL
url = 'http://web.mta.info/developers/turnstile.html'  #this is the main url that consists the list of urls that
                                                #contain the link to the txt files
```

```python

# Send request and verify URL
rsp = requests.get(url) #must send a requests to get the url
print(rsp.url)
```

```python

# Parse HTML response
soup = BeautifulSoup(rsp.text, 'html.parser') #this will parse out the html response
print(soup)
```

```python

# Find all anchor tags with href attribute defined
a_tags = soup.find_all('a', href=True)  #Find all 'a' tags

file_list = []
keys = []
newkeys = []
base_url = url.split('turnstile')[0] #Takes everything from left of the turnstile delimiter, this is the base url
for a in a_tags:
    tmp = re.search('data/nyct/turnstile/turnstile_\d*\.txt', a['href'])
    if tmp is not None:
        file_list.append(''.join([base_url, tmp.string]))
for item in file_list:
    key = int((item.split('_')[1]).split('.')[0])
    keys.append(key)
print(keys)
print(file_list)
for key1 in keys:
    if 150307<=key1<=150606 or 160305 <=key1<=160604 or 170304 <= key1 <= 170603:
        newkeys.append(key1)
print(newkeys)
print(file_list)
```
