```python
# Create function to find Ticket Revenue (live gate)

def TicketRev(soup):
    #Must use try and except b/c some tables are missing ticket revenue
     try:
        for ultag in soup.find_all('ul',{'class':'clearfix'}):  #ticket revenue is found in list called class clearfix
            for litag in ultag.find_all('li'):
                soup2 = litag.find(text="Ticket Revenue (live gate):")
                if soup2 is not None:
                    soup3 = soup2.findNext().text
        #remove $ and , sign to make the ticket revenue into integer
        ticketrev = soup3.replace(',','')
        ticketrev = ticketrev.replace('$','')
        if "M" in ticketrev:  #convert M (million sign) into integer
            ticketrev = float(ticketrev.replace('M',''))*1e6
        return int(ticketrev)
     except:
        pass
```

```python
#Create function to find total attendance
def TotAttend(soup):
     try:
        for ultag in soup.find_all('ul',{'class':'clearfix'}):
            for litag in ultag.find_all('li'):
                soup2 = litag.find(text="Attendance:")
                if soup2 is not None:
                    soup3 = soup2.findNext().text
        
        attend = soup3.replace(',','')
        if "M" in attend:
            attend = float(attend.replace('M',''))*1e6
        return int(attend)
     except:
        pass
```

```python
def PPV(soup):
    try:
        for ultag in soup.find_all('ul',{'class':'clearfix'}):
            for litag in ultag.find_all('li'):
                soup2 = litag.find(text="PPV Buys / Buyrate:")
                if soup2 is not None:
                    soup3 = soup2.findNext().text
        
        ppv = soup3.replace(',','')
        if "M" in ppv:
            ppv = float(ppv.replace('M',''))*1e6
        return int(ppv)
    except:
        pass
```

```python
def TVRate(soup):
      try:
        for ultag in soup.find_all('ul',{'class':'clearfix'}):
            for litag in ultag.find_all('li'):
                soup2 = litag.find(text="TV Ratings:")
                if soup2 is not None:
                    soup3 = soup2.findNext().text.split(' ')[0] #we are interested in the first position b/c tv rating is followed by other text
                    
        
        tv = soup3.replace(',','')
        if "M" in tv:
            tv = float(tv.replace('M',''))*1e6
        return int(tv)
      except:
        pass
```

```python
#Create a function that will parse out the html code, create key for dictionary, and if the main event featured a
#rematch trilogy 
def parse_html_hits(url):
    response_t = requests.get(url)
    page_t = response_t.text
    soup_t = BeautifulSoup(page_t,'lxml')
    rematch = 0
    
    key = soup_t.find_all(class_ = 'b-content__title-highlight')
    key1 = str(key).split('>')[1].replace('\n','').strip().split(':')[0].strip('</span').strip()
    key2 = str(key).split('>')[1].replace('\n','').replace('</span','').strip()
    m = re.search(r'\d+$', key2)  #if the main event contains a number it must be their second fight, therefore it's counted as a rival
    if m is not None:
        rematch = 1
    return soup_t,key1,rematch
```

```python
#Create function to add up all strikes in the UFC event
def totStrikes(soup_t1):
    table_t1 =soup_t1.find_all('table')
    td1 = table_t1[0].find_all('td')
    i = 2
    totstr = 0
    while i <= (len(td1)-8):
        td2 = table_t1[0].find_all('td')[i]
        td3 = td2.text.split()
        for s in td3:
            totstr += int(s)
        i = i + 10
    return int(totstr)
```

```python
#Create function to add up all take downs in a table
def totTD(soup_t1):
    table_t1 =soup_t1.find_all('table')
    td1 = table_t1[0].find_all('td')
    i = 3
    tottd = 0
    while i <= (len(td1)-7):
        td2 = table_t1[0].find_all('td')[i]
        td3 = td2.text.split()
        for s in td3:
            tottd += int(s)
        i = i + 10
    return tottd
```

```python
#Create a function to add up all submissions in a table 
def totSub(soup_t1):
    table_t1 =soup_t1.find_all('table')
    td1 = table_t1[0].find_all('td')
    i = 4
    totsub = 0
    while i <= (len(td1)-6):
        td2 = table_t1[0].find_all('td')[i]
        td3 = td2.text.split()
        for s in td3:
            totsub += int(s)
        i = i + 10
    return totsub
```

```python
#Create a function to add up all passes in a table
def totPasses(soup_t1):
    table_t1 =soup_t1.find_all('table')
    td1 = table_t1[0].find_all('td')
    i = 5
    totpass = 0
    while i <= (len(td1)-5):
        td2 = table_t1[0].find_all('td')[i]
        td3 = td2.text.split()
        for s in td3:
            totpass += int(s)
        i = i + 10
    return totpass
```

```python
#Create a function that will create a list of weight classes and if a belt was on the line
def WeightClass(soup_t1):
    table_t1 =soup_t1.find_all('table')
    td1 = table_t1[0].find_all('td')
    i = 6
    lb_class = []
    belt = 0
    light = 0
    while i <= (len(td1)-4):
        td2 = table_t1[0].find_all('td')[i]
        weight = td2.text.replace('\n','').strip()
        lb_class.append(weight)
        if td2.img is not None:
            if "belt" in str(td2.img):
                belt += 1
        i = i + 10
    wg_class = ['Strawweight','Flyweight','Bantamweight','Featherweight','Lightweight','Super lightweight',
               "Women's Strawweight","Women's Flyweight","Women's Bantamweight","Women's Featherweight"]
    women_list = ["Women's Strawweight","Women's Flyweight","Women's Bantamweight","Women's Featherweight"]
    women = 0
    #Get a count of lighter weight and heavy weight fights
    for weight in lb_class:
        if weight in wg_class:
            light += 1
        if weight in women_list:
            women += 1
    heavy = len(lb_class) - light
    men = len(lb_class) - women
    all_list = [light,heavy,women,men,belt]
    return all_list
```

```python
#Create function that will create a list of KO, U-Dec,and Other
def fightEnd(soup_t1):
    table_t1 =soup_t1.find_all('table')
    td1 = table_t1[0].find_all('td')
    i = 7
    end = []
    KO = 0
    DEC = 0
    SUB = 0
    other = 0
    while i <= (len(td1)-3):
        td2 = table_t1[0].find_all('td')[i]
        end1 = td2.text.replace('\n','').strip()
        end.append(end1)
        i = i + 10
    for end1 in end:
        if 'KO' in end1:
            KO += 1
        elif 'DEC' in end1:
            DEC += 1
        elif 'SUB' in end1:
            SUB += 1
        else:
            other += 1
    all_list = [KO,DEC,SUB,other]
    return all_list
```

```python
#Create function to count the total number of rounds and total time of fights
import time
from datetime import datetime
def countRoundsTime(soup_t1):
    table_t1 =soup_t1.find_all('table')
    td1 = table_t1[0].find_all('td')
    i = 8
    rds = 0
    rd_list= []
    t_list2 = []
    sec_total = 0
    while i <= (len(td1)-2):
        td2 = table_t1[0].find_all('td')[i]
        rd = int(td2.text.replace('\n','').strip())
        rd_list.append(rd)
        rds += rd
        i = i + 10
  
    j = 9
    t_list = []
    while j <= (len(td1)-1):
        td3 = table_t1[0].find_all('td')[j]
        t = td3.text.replace('\n','').strip()
        t_list.append(t)
        j = j + 10
    for t in t_list:
        m, s = t.split(':')
        sec = int(m) * 60 + int(s)
        t_list2.append(sec)
    for (x,y) in zip(rd_list,t_list2):
        secs = x * y
        sec_total += secs
#     totmin = sec_total // 60
#     totsec = sec_total % 60
    m, s = divmod(sec_total, 60)
    h, m = divmod(m, 60)
    tottime = "%d:%02d:%02d" % (h, m, s)
    r = [rds,tottime,sec_total]
    
    return r
```

```python
#Create a function to bring all the functions together to create dictionary and transform into dataframe, fight # is the key and all the
#info are the values
def Table2(url):
    soup = parse_html_hits(url)[0]  
    rematch = parse_html_hits(url)[2]
    strikes = totStrikes(soup)
    
    takedowns = totTD(soup)
    submission = totSub(soup)
    passes = totPasses(soup)
    weightClass = WeightClass(soup)
    lightfights = weightClass[0]
    heavyfights = weightClass[1]
    women = weightClass[2]
    men = weightClass[3]
    belts = weightClass[4]
    end = fightEnd(soup)
    KO = end[0]
    DEC = end[1]
    SUB = end[2]
    Other = end[3]
    time = countRoundsTime(soup)
    rounds = time[0]
    tottime = time[2]
    key = parse_html_hits(url)[1]
    superstar = SuperStar(soup)[0]
    rival = SuperStar(soup)[1]
    conor = SuperStar(soup)[2]
    jones = SuperStar(soup)[3]
    rousey = SuperStar(soup)[4]
    silva = SuperStar(soup)[5]
    pierre = SuperStar(soup)[6]
    lesnar = SuperStar(soup)[7]
    liddell = SuperStar(soup)[8]
    dictionary = {} #Create dictionary to transform into dataframe
    dictionary[key] = [key,strikes,takedowns,submission,passes,lightfights,heavyfights,women,men,belts,KO,DEC,SUB,Other,rematch,superstar,rival,conor,jones,rousey,silva,pierre,lesnar,liddell,rounds,tottime]
    dict_to_df = pd.DataFrame.from_dict(dictionary,orient='index')
    dict_to_df.columns = ['fight','strikes', 'takedowns','submission','passes','lightfights','heavyfights','women','men','belts','KO','DEC','SUB','Other','rematch','superstar','rival','conor','jones','rousey','silva','pierre','lesnar','liddell','rounds','tottime']

    return dict_to_df
```
