```python
#Function returns ppv buys and tv ratings
def PPV_TVRat(soup):
    try:
        obj = soup[1].find_all('li')[13]
        ppv = obj.find(text="PPV Buys / Buyrate:").parent.parent.findNext('span').contents[0]
        ppv = int(ppv.replace(',',''))

        tv = obj.find(text="TV Ratings:").parent.findNext('span').contents[0]
        tv = tv.split(' ')[0].replace(",","")
        if "M" in tv:
           
            tv = int(float(tv.replace("M",'')) * 1e6)
            
        elif "M" not in tv:
            tv = int(tv)
        return ppv,tv
    except:
        pass
```

```python
#Function returns parsed out html code that will be able to be fed into each function 
def parsed_html(url):
#     url2 = 'http://www.tapology.com/fightcenter/events/47255-ufc-218'
    response = requests.get(url)
    page = response.text
    soup = BeautifulSoup(page,'lxml')
    soup2 = soup.find_all(class_ = 'right')
    return soup,soup2
```

```python
#Bring all the functions together to make a dataframe
import pandas as pd
import numpy as np

def fightInfo(url):
    try:
        
        soup = parsed_html(url)[1]
        soup2 = parsed_html(url)[0]
        soup3 = soup2.find_all(class_="selected")
        for word in [i.text for i in soup3]:
            key = word.strip()
        
        bouts = MMA_Bouts(soup)
        ppv = PPV_TVRat(soup)[0]
        tvrate = PPV_TVRat(soup)[1]
        attend = TotAttend(soup)
        ticketrev = TicketRev(soup)
        dictionary = {}
        dictionary[key] = [key,bouts,tvrate,attend,ticketrev,ppv]
        dict_to_df = pd.DataFrame.from_dict(dictionary,orient='Index')
        dict_to_df.columns = ['fight','bouts','tvrate','attend','ticketrev','ppv']

        return dict_to_df
    except:
        pass
    
    
    
```

```python
def parse_html_hits(url):
    response_t = requests.get(url)
    page_t = response_t.text
    soup_t = BeautifulSoup(page_t,'lxml')
    rival = 0
    key = soup_t.find_all(class_ = 'b-content__title-highlight')
    key1 = str(key).split('>')[1].replace('\n','').strip().split(':')[0]
    key2 = str(key).split('>')[1].replace('\n','').replace('</span','').strip()
    
    m = re.search(r'\d+$', key2)
    if m is not None:
        rival = 1
    return soup_t,key1,rival
```

```python
#Create function to add up all strikes in a table
def totStrikes(soup_t1):
    table_t1 =soup_t1.find_all('table')
    td1 = table_t1[0].find_all('td')
    i = 2
    totstr = 0
    while i <= (len(td1)-8):
        td2 = table_t1[0].find_all('td')[i]
        td3 = td2.text.split()
        for s in td3:
            totstr += int(s)
        i = i + 10
    return int(totstr)
```

```python
#Create a function that will create a list of weight classes and if a belt was on the line
def WeightClass(soup_t1):
    table_t1 =soup_t1.find_all('table')
    td1 = table_t1[0].find_all('td')
    i = 6
    lb_class = []
    belt = 0
    light = 0
    while i <= (len(td1)-4):
        td2 = table_t1[0].find_all('td')[i]
        weight = td2.text.replace('\n','').strip()
        lb_class.append(weight)
        if td2.img is not None:
            if "belt" in str(td2.img):
                belt += 1
        i = i + 10
    wg_class = ['Strawweight','Flyweight','Bantamweight','Featherweight','Lightweight','Super lightweight',
               "Women's Strawweight","Women's Flyweight","Women's Bantamweight","Women's Featherweight"]
    women_list = ["Women's Strawweight","Women's Flyweight","Women's Bantamweight","Women's Featherweight"]
    women = 0
    #Get a count of lighter weight and heavy weight fights
    for weight in lb_class:
        if weight in wg_class:
            light += 1
        if weight in women_list:
            women += 1
    heavy = len(lb_class) - light
    men = len(lb_class) - women
    all_list = [light,heavy,women,men,belt]
    return all_list
```

```python
#Create function that will create a list of KO, U-Dec,CNC
def fightEnd(soup_t1):
    table_t1 =soup_t1.find_all('table')
    td1 = table_t1[0].find_all('td')
    i = 7
    end = []
    KO = 0
    DEC = 0
    SUB = 0
    other = 0
    while i <= (len(td1)-3):
        td2 = table_t1[0].find_all('td')[i]
        end1 = td2.text.replace('\n','').strip()
        end.append(end1)
        i = i + 10
    for end1 in end:
        if 'KO' in end1:
            KO += 1
        elif 'DEC' in end1:
            DEC += 1
        elif 'SUB' in end1:
            SUB += 1
        else:
            other += 1
    all_list = [KO,DEC,SUB,other]
    return all_list
```

```python
#Create function to count the number of rounds
import time
from datetime import datetime
def countRoundsTime(soup_t1):
    table_t1 =soup_t1.find_all('table')
    td1 = table_t1[0].find_all('td')
    i = 8
    rds = 0
    rd_list= []
    t_list2 = []
    sec_total = 0
    while i <= (len(td1)-2):
        td2 = table_t1[0].find_all('td')[i]
        rd = int(td2.text.replace('\n','').strip())
        rd_list.append(rd)
        rds += rd
        i = i + 10
  
    j = 9
    t_list = []
    while j <= (len(td1)-1):
        td3 = table_t1[0].find_all('td')[j]
        t = td3.text.replace('\n','').strip()
        t_list.append(t)
        j = j + 10
    for t in t_list:
        m, s = t.split(':')
        sec = int(m) * 60 + int(s)
        t_list2.append(sec)
    for (x,y) in zip(rd_list,t_list2):
        secs = x * y
        sec_total += secs
#     totmin = sec_total // 60
#     totsec = sec_total % 60
    m, s = divmod(sec_total, 60)
    h, m = divmod(m, 60)
    tottime = "%d:%02d:%02d" % (h, m, s)
    r = [rds,tottime]
    return r
```

```python
def Table2(url):
    soup = parse_html_hits(url)[0]
    rival = parse_html_hits(url)[2]
    strikes = totStrikes(soup)
    
    takedowns = totTD(soup)
    submission = totSub(soup)
    passes = totPasses(soup)
    weightClass = WeightClass(soup)
    lightfights = weightClass[0]
    heavyfights = weightClass[1]
    women = weightClass[2]
    men = weightClass[3]
    belts = weightClass[4]
    end = fightEnd(soup)
    KO = end[0]
    DEC = end[1]
    SUB = end[2]
    Other = end[3]
    time = countRoundsTime(soup)
    rounds = time[0]
    tottime = time[1]
    key = parse_html_hits(url)[1]
    dictionary = {}
    dictionary[key] = [key,strikes,takedowns,submission,passes,lightfights,heavyfights,women,men,belts,KO,DEC,SUB,Other,rival,rounds,tottime]
    dict_to_df = pd.DataFrame.from_dict(dictionary,orient='index')
    dict_to_df.columns = ['fight','strikes', 'takedowns','submission','passes','lightfights','heavyfights','women','men','belts','KO','DEC','SUB','Other','rival','rounds','tottime']

    return dict_to_df
```
